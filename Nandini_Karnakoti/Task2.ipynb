## **1. Importing Libraries**
Start by importing the necessary libraries.


```python
import pandas as pd
```

 ## **2. Loading Excel File**

Here, I have loaded the Excel file named "Sales" containing the data.


```python
file_path = '/content/Sales Data.xlsx'
excel_file = pd.ExcelFile(file_path)
```

## **3. Raw Sales Data**

Previewing the raw **Sales** sheet before cleaning.


```python
sales_data = pd.read_excel(file_path, sheet_name='Sales')
print("Sales Data (Before Cleaning):\n")
sales_data.head()
```

## **4. Sales Data Cleaning**
Cleaning the **Sales** sheet by removing unnecessary rows and columns, renaming columns,
and ensuring the 'Period' column is in datetime format.

```python
sales_data_cleaned = sales_data.iloc[1:].reset_index(drop=True)
sales_data_cleaned.columns = ['Period', 'Value_Brand', 'V1', 'V2', 'V3',
                              'Unused', 'Volume_Brand', 'V1_Volume', 'V2_Volume', 'V3_Volume',
                              'Unused_2', 'Avg_Price_per_SU', 'V1_Avg_Price',
                              'V2_Avg_Price', 'V3_Avg_Price', 'Unused_3',
                              'Avg_Price_per_USD', 'V1_USD', 'V2_USD', 'V3_USD']
sales_data_cleaned['Period'] = pd.to_datetime(sales_data_cleaned['Period'], errors='coerce')
sales_data_cleaned = sales_data_cleaned.drop(columns=['Unused', 'Unused_2', 'Unused_3'])
print("Sales Data (After Cleaning):\n")
print(sales_data_cleaned.head().to_string(index=False))
```

## **5. Raw Distribution Data**



Previewing the raw **Distribution** sheet before cleaning.

```python
distribution_data = pd.read_excel(file_path, sheet_name='Distribution')
print("Distribution Data (Before Cleaning):\n")
distribution_data.head()
```

## **6. Distribution Data Cleaning**

Next, cleaning the **Distribution** sheet by following similar steps.

```python
distribution_data_cleaned = distribution_data.iloc[1:].reset_index(drop=True)
distribution_data_cleaned.columns = ['Period', 'WTD_Brand', 'V1_WTD', 'V2_WTD', 'V3_WTD',
                                      'Unused', 'NUMD_Brand', 'V1_NUMD', 'V2_NUMD', 'V3_NUMD']
distribution_data_cleaned['Period'] = pd.to_datetime(distribution_data_cleaned['Period'], errors='coerce')
distribution_data_cleaned = distribution_data_cleaned.drop(columns=['Unused'])
print("Distribution Data (After Cleaning):\n")
print(distribution_data_cleaned.head().to_string(index=False))
```

## **7. Raw Shipment Data**

Previewing the raw **Shipment** sheet before cleaning.


```python
shipment_data = pd.read_excel(file_path, sheet_name='Shipment')
print("Shipment Data (Before Cleaning):\n")
shipment_data.head()
```

## **8. Shipment Data Cleaning**

Cleaning the **Shipment** sheet similarly.


```python
shipment_data_cleaned = shipment_data.iloc[1:].reset_index(drop=True)
shipment_data_cleaned.columns = ['Period', 'Brand_Shipment', 'V1_Shipment', 'V2_Shipment',
                                  'V3_Shipment', 'Unused', 'Volume_Brand', 'V1_Volume',
                                  'V2_Volume', 'V3_Volume']
shipment_data_cleaned['Period'] = pd.to_datetime(shipment_data_cleaned['Period'], errors='coerce')
shipment_data_cleaned = shipment_data_cleaned.drop(columns=['Unused'])
print("Shipment Data (After Cleaning):\n")
print(shipment_data_cleaned.head().to_string(index=False))
```

```python
# Merging sales and distribution data
merged_data = pd.merge(sales_data_cleaned, distribution_data_cleaned, on='Period', how='outer')

# Merging the result with shipment data
final_merged_data = pd.merge(merged_data, shipment_data_cleaned, on='Period', how='outer')

print("Final Merged Data:\n")
print(final_merged_data.head().to_string(index=False))
```

```python
# Concatenate the three DataFrames along columns
combined_data = pd.concat([sales_data_cleaned, distribution_data_cleaned, shipment_data_cleaned], axis=1)

# Drop duplicate columns
combined_data = combined_data.loc[:, ~combined_data.columns.duplicated()]

print("Combined Data (After Merging Columns):\n")
print(combined_data.head().to_string(index=False))
```
